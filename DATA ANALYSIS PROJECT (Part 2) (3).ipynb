# Data Visualisation

### Univariate Analysis

```python
plt.figure(figsize = (16, 8))
myexplode = [0]*12
myexplode[0] = 0.3
palette_r = sns.color_palette("Blues", 15)

plt.pie(df1["Brands"].value_counts(), labels = sorted(list(df1["Brands"].unique())), explode = myexplode, colors = palette_r, autopct='%.0f%%');

plt.savefig("Pie Chart", bbox_inches='tight', format = 'jpg', dpi = 600)
plt.title("Model distributions of Air Conditioners in each brand")
plt.show()
```

### Bivariate Analysis

```python
# Range of prices that exist in each brand
plt.figure(figsize=(20,12))
ax = sns.barplot(x = df1["Brands"] , y = df1["Prices"] , ci = None);

for i in ax.containers:
    ax.bar_label(i,padding=5)
    
plt.show()
```

```python
# Brands producing different Star ACs
plt.figure(figsize=(20,12))
ax = sns.barplot(x = df1["Brands"] , y = df1["Stars"] , ci = None);

for i in ax.containers:
    ax.bar_label(i,padding=5)
    
plt.show
```

# Box Plot

### Monthly

```python
box_plot_data = [ np.array(df1["Monthly Charges(Without Solar Panels)"]),np.array(df1["Monthly Charges(With 3 Solar Panels)"]),np.array(df1["Monthly Charges(With 5 Solar Panels)"]),np.array(df1["Monthly Charges(With 7 Solar Panels)"])]
box_plot_data
lbl = ['Monthly Charges(Without Solar Panels)','Monthly Charges(With 3 Solar Panels)','Monthly Charges(With 5 Solar Panels)','Monthly Charges(With 7 Solar Panels)']
plt.figure(figsize=(20,10))
plt.boxplot(box_plot_data, labels=lbl);
```

```python
box_plot_data = [ np.array(df1["5 Years' Charges(Without Solar Panels)"]),np.array(df1["5 Years' Charges(With 3 Solar Panels)"]),np.array(df1["5 Years' Charges(With 5 Solar Panels)"]),np.array(df1["5 Years' Charges(With 7 Solar Panels)"])]
box_plot_data
lbl = ["5 Years' Charges(Without Solar Panels)","5 Years' Charges(With 3 Solar Panels)","5 Years' Charges(With 5 Solar Panels)","5 Years' Charges(With 7 Solar Panels)"]
plt.figure(figsize=(20,10))
plt.boxplot(box_plot_data, labels=lbl);
```
