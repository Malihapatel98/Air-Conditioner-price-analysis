# Extracting data from all the URLs and making the final dataframe

```python
url = """https://www.flipkart.com/search?q=AC&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off="""

brands = []
types = []
watts = []
capacities = []
noise_level = []
stars = []
prices = []
ratings = []

for i in range(1,33):
    urls = url+str(i)
    print(urls)
    page = requests.get(urls)
    soup = BeautifulSoup(page.text)
    
    for i in price:
        a = i.text
        a1 = a.strip("₹").replace("," , "")
        prices.append(int(a1)) 
    
    for i in brand:
        a = i.text
        a1 = a.split(" ")

        if a1[0] == "Blue":
            brands.append(a1[0] + " " + a1[1])
        else:
            brands.append(a1[0])
            
    for i in watt:
        a = i.text
        b = a.replace(": ", "")
        x = re.findall("Power Consumption([\W\s\d+\w+]*)" , b)
        if x:
            watts.append(x[0])
        
    for i in type1:
        a = i.text
        d = re.findall(r"Split|Window" , a)
        types.append(d[0])
    
    for i in capacity:
        a = i.text
        b = re.findall(r"\d+.\d+\sTon | \d+\sTon",a)
        capacities.append(b[0])      
    
    for i in stars1:
        a = i.text
        x = re.findall(r"\d+\sStar" , a)
        stars.append(x[0])
    
    for i in rating:
        a = i.text
        ratings.append(a)
        
    for i in decibels:
        a = i.text
        b = a.replace(": ", "")
        x = re.findall("Noise level([\W\s\d+\w+]*)" , b)
        if x:
            noise_level.append(x[0])
            
    dictionary = {"Brands":brands[0:23], "Types" : types[0:23], "Capacity" : capacities[0:23], "Power Consumption" : watts[0:23], "Noise Level" : noise_level[0:23], "Stars" : stars[0:23], "Prices" : prices[0:23], "Rating" : ratings[0:700]}

df1 = pd.DataFrame(dictionary)
df1
```

```python
df1["Capacity"] = df1["Capacity"].astype("float")
df1["Noise Level"] = df1["Noise Level"].astype("float")
df1["Stars"] = df1["Stars"].astype("int")
df1["Rating"] = df1["Rating"].astype("float")
df1["Power Consumption"] = df1["Power Consumption"].astype("float")
```

```python
list1 = []

for x in df1["Power Consumption"]:
    ac_power_consumption = (x * 8 * 30) / 1000
    if (ac_power_consumption <= 0):
        list1.append(ac_power_consumption * 0)
    elif (ac_power_consumption > 0) and (ac_power_consumption < 201):
        list1.append(ac_power_consumption * 5)
    elif(ac_power_consumption >= 201) and (ac_power_consumption < 300):
        list1.append(ac_power_consumption * 7.2)
    elif(ac_power_consumption >= 301) and (ac_power_consumption < 400):
        list1.append(ac_power_consumption * 8.5)
    elif(ac_power_consumption >= 401) and (ac_power_consumption < 800):
        list1.append(ac_power_consumption * 9)
    else:
        list1.append(ac_power_consumption * 9.5)
        
        
df1["Monthly Charges(Without Solar Panels)"] = list1  

list2 = []

for x in df1["Power Consumption"]:
    ac_power_consumption = (x * 8 * 30) / 1000
    if (ac_power_consumption > 0) and (ac_power_consumption < 201):
        list2.append((ac_power_consumption * 5) * 60)
    elif(ac_power_consumption >= 201) and (ac_power_consumption < 300):
        list2.append((ac_power_consumption * 7.2) * 60)
    elif(ac_power_consumption >= 301) and (ac_power_consumption < 400):
        list2.append((ac_power_consumption * 8.5) * 60)
    elif(ac_power_consumption >= 401) and (ac_power_consumption < 800):
        list2.append((ac_power_consumption * 9) * 60)
    else:
        list2.append((ac_power_consumption * 9.5) * 60)
df1["5 Years' Charges(Without Solar Panels)"] = list2

list3 = []

for x in df1["Power Consumption"]:
    ac_power_consumption = (x * 8 * 30) / 1000
    solar_power_production = 3 * ((260 * 8 * 30) / 1000)         # For 3 Solar Panels
    difference = ac_power_consumption - solar_power_production
    if (difference <= 0):
        list3.append(0)
    elif (difference > 0) and (difference < 201):
        list3.append(difference * 5)
    elif(difference >= 201) and (difference < 300):
        list3.append(difference * 7.2)
    elif(difference >= 301) and (difference < 400):
        list3.append(difference * 8.5)
    elif(difference >= 401) and (difference < 800):
        list3.append(difference * 9)
    else:
        list3.append(difference * 9.5)
        

df1["Monthly Charges(With 3 Solar Panels)"] = list3

list4 = []

for x in df1["Power Consumption"]:
    ac_power_consumption = (x * 8 * 30) / 1000
    solar_power_production = 5 * ((260 * 8 * 30) / 1000)         # For 5 Solar Panels
    difference = ac_power_consumption - solar_power_production
    if (difference <= 0):
        list4.append(0)
    elif (difference > 0) and (difference < 201):
        list4.append(difference * 5)
    elif(difference >= 201) and (difference < 300):
        list4.append(difference * 7.2)
    elif(difference >= 301) and (difference < 400):
        list4.append(difference * 8.5)
    elif(difference >= 401) and (difference < 800):
        list4.append(difference * 9)
    else:
        list4.append(difference * 9.5)
        

df1["Monthly Charges(With 5 Solar Panels)"] = list4

list5 = []

for x in df1["Power Consumption"]:
    ac_power_consumption = (x * 8 * 30) / 1000
    solar_power_production = 7 * ((260 * 8 * 30) / 1000)         # For 7 Solar Panels
    difference = ac_power_consumption - solar_power_production
    if (difference <= 0):
        list5.append(0)
    elif (difference > 0) and (difference < 201):
        list5.append(difference * 5)
    elif(difference >= 201) and (difference < 300):
        list5.append(difference * 7.2)
    elif(difference >= 301) and (difference < 400):
        list5.append(difference * 8.5)
    elif(difference >= 401) and (difference < 800):
        list5.append(difference * 9)
    else:
        list5.append(difference * 9.5)
        

df1["Monthly Charges(With 7 Solar Panels)"] = list5

list6 = []

for x in df1["Power Consumption"]:
    ac_power_consumption = (x * 8 * 30) / 1000
    solar_power_production = 3 * ((260 * 8 * 30) / 1000)         # For 3 Solar Panels
    difference = ac_power_consumption - solar_power_production
    if (difference <= 0):
        list6.append(0)
    elif (difference > 0) and (difference < 201):
        list6.append((difference * 5) * 60)
    elif(difference >= 201) and (difference < 300):
        list6.append((difference * 7.2) * 60)
    elif(difference >= 301) and (difference < 400):
        list6.append((difference * 8.5) * 60)
    elif(difference >= 401) and (difference < 800):
        list6.append((difference * 9) * 60)
    else:
        list6.append((difference * 9.5) * 60)
        

df1["5 Years' Charges(With 3 Solar Panels)"] = list6

list7 = []

for x in df1["Power Consumption"]:
    ac_power_consumption = (x * 8 * 30) / 1000
    solar_power_production = 5 * ((260 * 8 * 30) / 1000)         # For 5 Solar Panels
    difference = ac_power_consumption - solar_power_production
    if (difference <= 0):
        list7.append(0)
    elif (difference > 0) and (difference < 201):
        list7.append((difference * 5) * 60)
    elif(difference >= 201) and (difference < 300):
        list7.append((difference * 7.2) * 60)
    elif(difference >= 301) and (difference < 400):
        list7.append((difference * 8.5) * 60)
    elif(difference >= 401) and (difference < 800):
        list7.append((difference * 9) * 60)
    else:
        list7.append((difference * 9.5) * 60)
        

df1["5 Years' Charges(With 5 Solar Panels)"] = list7

list8 = []

for x in df1["Power Consumption"]:
    ac_power_consumption = (x * 8 * 30) / 1000
    solar_power_production = 7 * ((260 * 8 * 30) / 1000)         # For 7 Solar Panels
    difference = ac_power_consumption - solar_power_production
    if (difference <= 0):
        list8.append(0)
    elif (difference > 0) and (difference < 201):
        list8.append((difference * 5) * 60)
    elif(difference >= 201) and (difference < 300):
        list8.append((difference * 7.2) * 60)
    elif(difference >= 301) and (difference < 400):
        list8.append((difference * 8.5) * 60)
    elif(difference >= 401) and (difference < 800):
        list8.append((difference * 9) * 60)
    else:
        list8.append((difference * 9.5) * 60)
        

df1["5 Years' Charges(With 7 Solar Panels)"] = list8

df1
```

# Rounding off the decimals upto 2 values

```python
monthly_charges_without_solar_panels = []
five_years_charges_without_solar_panels = []
monthly_charges_with_3_solar_panels = []
monthly_charges_with_5_solar_panels = []
monthly_charges_with_7_solar_panels = []
five_years_charges_with_3_solar_panels = []
five_years_charges_with_5_solar_panels = []
five_years_charges_with_7_solar_panels = []


for x in df1["Monthly Charges(Without Solar Panels)"]:
    monthly_charges_without_solar_panels.append(round(x , 2))
    
for x in df1["5 Years' Charges(Without Solar Panels)"]:
    five_years_charges_without_solar_panels.append(round(x , 2))
    
for x in df1["Monthly Charges(With 3 Solar Panels)"]:
    monthly_charges_with_3_solar_panels.append(round(x , 2))
    
for x in df1["Monthly Charges(With 5 Solar Panels)"]:
    monthly_charges_with_5_solar_panels.append(round(x , 2))
    
for x in df1["Monthly Charges(With 7 Solar Panels)"]:
    monthly_charges_with_7_solar_panels.append(round(x , 2))
    
for x in df1["5 Years' Charges(With 3 Solar Panels)"]:
    five_years_charges_with_3_solar_panels.append(round(x , 2))
    
for x in df1["5 Years' Charges(With 5 Solar Panels)"]:
    five_years_charges_with_5_solar_panels.append(round(x , 2))
    
for x in df1["5 Years' Charges(With 7 Solar Panels)"]:
    five_years_charges_with_7_solar_panels.append(round(x , 2))
    
df1["Monthly Charges(Without Solar Panels)"] = monthly_charges_without_solar_panels
df1["5 Years' Charges(Without Solar Panels)"] = five_years_charges_without_solar_panels
df1["Monthly Charges(With 3 Solar Panels)"] = monthly_charges_with_3_solar_panels
df1["Monthly Charges(With 5 Solar Panels)"] = monthly_charges_with_5_solar_panels
df1["Monthly Charges(With 7 Solar Panels)"] = monthly_charges_with_7_solar_panels
df1["5 Years' Charges(With 3 Solar Panels)"] = five_years_charges_with_3_solar_panels
df1["5 Years' Charges(With 5 Solar Panels)"] = five_years_charges_with_5_solar_panels
df1["5 Years' Charges(With 7 Solar Panels)"] = five_years_charges_with_7_solar_panels
```
